```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simple movie dataset
movies = {
    "Title": ["Inception", "Interstellar", "The Dark Knight", "Avatar", "Avengers", "Titanic"],
    "Genre": ["Sci-Fi", "Sci-Fi", "Action", "Sci-Fi", "Action", "Romance"],
    "Rating": [8.8, 8.6, 9.0, 7.8, 8.0, 7.9],
    "Box_Office_M": [836, 701, 1006, 2923, 2799, 2264]
}

df = pd.DataFrame(movies)
print("--- Movies Dataset ---")
print(df)

# Setting styling for the chart
sns.set_theme(style="whitegrid")

# Chart: Average rating per genre
plt.figure(figsize=(5, 3))
sns.barplot(data=df, x="Genre", y="Rating", palette="pastel", errorbar=None)
plt.title("Average Rating by Movie Genre", fontweight="bold")
plt.ylim(5, 10)  # Starting Y axis at 5 to see variation clearly
plt.tight_layout()
plt.savefig("movie_rating.png")
plt.show()
```
